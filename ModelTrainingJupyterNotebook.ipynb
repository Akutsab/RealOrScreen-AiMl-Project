```python
results2 = model.train(
    data="dataset",
    epochs=80,
    imgsz=224,
    batch=8,

    optimizer="AdamW",
    lr0=5e-4,

    fliplr=0.5,
    flipud=0.0,

    degrees=3,
    translate=0.02,
    scale=0.05,

    hsv_h=0.005,
    hsv_s=0.05,
    hsv_v=0.05,

    mosaic=0.0,
    mixup=0.0,
    erasing=0.05,

    patience=20
)
```

Sample test case on a single image


```python
from ultralytics import YOLO

model = YOLO("/content/runs/classify/ScreenDetector/YOLOv8n/weights/best.pt")

result = model.predict(
    source="/content/57.jpg",
    imgsz=224,
    verbose=False
)[0]

# Predicted class
label = result.names[result.probs.top1]

# Confidence
confidence = float(result.probs.top1conf)

print("Predicted Label:", label)
print(f"Confidence: {confidence:.4f}")
```
